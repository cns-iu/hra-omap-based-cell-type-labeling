# Map OMAP Target Names to HGNC IDs.

```python
exp_cols = sorted(list(cell_expressions.columns))
display(exp_cols)
```

```python
omap_target_name_to_id_dict = {}

for i,row in omap.iterrows():
    target_name = row['target_name']
    target_hgnc_id = row['HGNC_ID']

    # Handle naming discrepancies
    if target_name == 'Alpha smooth muscle actin':
        target_name = 'aSMA'
    if target_name == 'Synaptophysin':
        target_name = 'Synapto'
    if target_name == 'MUC-1':
        target_name = 'MUC1'
    if target_name == 'MUC-2':
        target_name = 'MUC2'
    if target_name == 'BCL-2':
        target_name = 'BCL2'
    if target_name == 'MUC-1/EMA':
        target_name = 'MUC1'
    if target_name == 'Collagen IV':
        target_name = 'CollagenIV'
    if target_name == 'HLA-DR':
        target_name = 'HLADR'
    if target_name == 'NKG2D (CD314)':
        target_name = 'NKG2D'
    if target_name == 'Chromogranin A':
        target_name = 'CHGA'
    if target_name == 'Podoplanin':
        target_name = 'PDPN'
    if target_name == 'Defensin alpha 5':
        target_name = 'aDef5'
    if 'Cytokeratin' in target_name:
        cyto_names = target_name.split(',')
        cyto_ids = target_hgnc_id.split(',')

        target_name = 'Cytokeratin'
        omap_target_name_to_id_dict[target_name] = cyto_ids
        if target_name not in exp_cols:
            print("Unmatch! ", target_name)
        continue

    if target_name not in exp_cols:
        print("Unmatch! ", target_name)

    omap_target_name_to_id_dict[target_name] = target_hgnc_id

# Handle CK7 present in CODEX channel names.
omap_target_name_to_id_dict['CK7'] = 'HGNC:6445'
    
display(omap_target_name_to_id_dict)
```

# Map BPs to CT from ASCT_B Table

```python
bp_to_ct_asctb_dict = {}

for i,row in asctb.iterrows():
    bp_id_set = list(row[['BProtein/1/ID', 'BProtein/2/ID', 'BProtein/3/ID', 'BProtein/4/ID', 'BProtein/5/ID', 'BProtein/6/ID']].values)
    bp_id_set_nan_dropped  = []
    for val in bp_id_set:
        if not pd.isna(val):
            bp_id_set_nan_dropped.append(val)
    
    #NOTE: Might want to check key sequence before adding. Currently, (a,b) != (b,a) and will create a new key. Is this expected behaviour? 

    #Handle cases where ASCT+B table can have upto CT/3. In that case, take the highest CT.
    if "CT/3" in asctb.columns:
        key_to_check = tuple(bp_id_set_nan_dropped)
        if key_to_check in bp_to_ct_asctb_dict:
            curr_val = bp_to_ct_asctb_dict[key_to_check]
            value_to_add = [row['CT/3'], row['CT/3/LABEL'], row['CT/3/ID']]
            if value_to_add not in curr_val:
                curr_val.append(value_to_add)
                bp_to_ct_asctb_dict[tuple(bp_id_set_nan_dropped)] = curr_val
        else:
            value_to_add = [[row['CT/3'], row['CT/3/LABEL'], row['CT/3/ID']]]
            bp_to_ct_asctb_dict[tuple(bp_id_set_nan_dropped)] = value_to_add

    elif "CT/2" in asctb.columns:
        key_to_check = tuple(bp_id_set_nan_dropped)
        if key_to_check in bp_to_ct_asctb_dict:
            curr_val = bp_to_ct_asctb_dict[key_to_check]
            value_to_add = [row['CT/2'], row['CT/2/LABEL'], row['CT/2/ID']]
            if value_to_add not in curr_val:
                curr_val.append(value_to_add)
                bp_to_ct_asctb_dict[tuple(bp_id_set_nan_dropped)] = curr_val
        else:
            value_to_add = [[row['CT/2'], row['CT/2/LABEL'], row['CT/2/ID']]]
            bp_to_ct_asctb_dict[tuple(bp_id_set_nan_dropped)] = value_to_add

    else:
        key_to_check = tuple(bp_id_set_nan_dropped)
        if key_to_check in bp_to_ct_asctb_dict:
            curr_val = bp_to_ct_asctb_dict[key_to_check]
            value_to_add = [row['CT/1'], row['CT/1/LABEL'], row['CT/1/ID']]
            if value_to_add not in curr_val:
                curr_val.append(value_to_add)
                bp_to_ct_asctb_dict[tuple(bp_id_set_nan_dropped)] = curr_val
        else:
            value_to_add = [[row['CT/1'], row['CT/1/LABEL'], row['CT/1/ID']]]
            bp_to_ct_asctb_dict[tuple(bp_id_set_nan_dropped)] = value_to_add
            
```

```python
# Merge cell location and cell expression dataframes. Only keep the cells that are present in the expression dataframe.
cell_loc_exp = cell_expressions.merge(cell_locations, left_on='ID', right_on='ID', how='inner')
display(cell_loc_exp.head())
```

```python
# Define cell nuclear channel which will be ignored while computing max mean expression.
NUCLEAR_CHANNEL = "Hoechst1"
CHANNEL_TO_IGNORE_1 = "DRAQ5" # DRAQ5 is a cell permeable far-red fluorescent DNA dye; no DNA dyes are represented in the OMAP itself, but everyone uses one of the DNA dyes to stain nuclei in cells.
CHANNEL_TO_IGNORE_2 = "CD147" # OMAP-2 intestine represents healthy human not disease state; CD147 = approved symbol BSG = HGNC:1116 is disease state: CD147 regulates cell proliferation, apoptosis, and tumor cell migration, metastasis and differentiation, especially under hypoxic conditions.
```

```python
expression_columns = list(cell_loc_exp.columns)
expression_columns.remove("ID")
expression_columns.remove("x")
expression_columns.remove("y")
expression_columns.remove("z")
expression_columns.remove(NUCLEAR_CHANNEL)
expression_columns.remove(CHANNEL_TO_IGNORE_1)
expression_columns.remove(CHANNEL_TO_IGNORE_2)
print(expression_columns)
print(len(expression_columns))
```

```python
# max_column_per_row = cell_loc_exp[expression_columns].apply(lambda row: row.idxmax(), axis=1) # This gets only column with max value
# max_column_per_row = cell_loc_exp[expression_columns].apply(lambda row: row[row != 0].index.tolist(), axis=1) # This gets all columns with non-zero expression values
max_column_per_row = cell_loc_exp[expression_columns].apply(lambda row: row.nlargest(10).index.tolist(), axis=1) # This gets all columns with top-10 expression values
# Concatenate "ID" column with max_column_per_row
cell_target = pd.concat([cell_loc_exp['ID'], max_column_per_row], axis=1)
cell_target.columns = ['ID', 'target_name']
display(cell_target)
```

```python
def compare_lists(list1, list2):
    # Check if any values match
    any_match = any(item in list2 for item in list1)

    # Find the number of matching values
    matching_values = set(list1) & set(list2)
    num_matches = len(matching_values)
    # Find the list of values that don't match
    non_matching_values_1 = [item for item in list1 if item not in matching_values]
    non_matching_values_2 = [item for item in list2 if item not in matching_values]
    non_matching_values = non_matching_values_1 + non_matching_values_2


    return any_match, num_matches, non_matching_values
# Example usage
list1 = ['apple', 'banana', 'orange']
list2 = ('banana', 'grape', 'pear')

any_match, num_matches, non_matching_values = compare_lists(list1, list2)

print("Any match:", any_match)
print("Number of matches:", num_matches)
print("Non-Matching values:", non_matching_values)
```

```python
cell_loc_with_types_list = [] 

for i, row in cell_loc_with_target.iterrows():
    target_names = row['target_name'] #List of targets
    # cell_name_list = []
    # cell_id_list = []
    cell_match_list = []
    # print(target_name)
    try:
        if pd.isna(target_names):
            # data = {'x':row['x'], 'y':row['y'], 'z':row['z'], 'target_ids':target_names, 'cell_types':None,'cell_type_confidence':None}
            data = {'x':row['x'], 'y':row['y'], 'z':row['z'], 'cell_types':None,'number_of_cell_type_matches':None}
            cell_loc_with_types_list.append(data)
            continue
    except:
        pass

    # Handle case for Cytokeratin. Move HGNCID sublist to main list.
    contains_sublist = any(isinstance(element, list) for element in target_names)
    if contains_sublist:
        updated_list = []
        for element in target_names:
            if isinstance(element, list):
                updated_list.extend(element)
            else:
                updated_list.append(element)
        target_names = updated_list

    best_match = None
    best_match_level = -1
    conf_str = "None"
    best_key = []

    for key,value in bp_to_ct_asctb_dict.items():

        any_match, num_matches, non_matching_values = compare_lists(key, target_names)

        if any_match and num_matches > best_match_level:
            best_match = value
            best_match_level = num_matches
            best_key = key

    conf_str = str(best_match_level) + "/" + str(len(best_key))
    # data = {'x':row['x'], 'y':row['y'], 'z':row['z'], 'target_ids':target_names, 'cell_types':best_match,'cell_type_confidence':conf_str}
    data = {'x':row['x'], 'y':row['y'], 'z':row['z'], 'cell_types':best_match,'number_of_cell_type_matches':conf_str}
    cell_loc_with_types_list.append(data) 

```
